```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/iris.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
print("Dataset Information:")
df.info()

print("Checking for missing values:")
print(df.isnull().sum())
```

```python
print("Describe the dataset:")
df.describe()
```

```python
print("Class distribution counts:")
sns.countplot(x='species', data=df)
plt.title('Count of each species')
plt.show()
```

```python
# Feature Histograms
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

df[numeric_columns].hist(bins=20, figsize=(12, 8))
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
numeric_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(12, 8))

for idx, feature in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, idx)
    sns.violinplot(x="species", y=feature, data=df, inner="quartile", palette="muted")
    plt.title(f"Distribution of {feature} by Label", fontsize=12)
    plt.xlabel("species")
    plt.ylabel(feature)

plt.suptitle("Violin Plots of Features Grouped by species", fontsize=16)
plt.tight_layout()
plt.show()
```

#### Analysis

Classes are balanced. Hence no augmentation is needed.

#### Histogram Analysis:

High separability = easy classification = higher accuracy

We don't need heavy feature engineering — features are already powerful.

#### Voilin Plot Analysis:

Petal Length and Petal Width are the most discriminative features.

Sepal Length is moderately useful, mainly for distinguishing Setosa.

Sepal Width has poor separability and may contribute noise if used alone.
